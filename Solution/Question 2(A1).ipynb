## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
import numpy as np
import sys
import math
np.set_printoptions(threshold=sys.maxsize)
```

## Reading Data from File

```python
#Reading data from ex1data1.txt file
data=pd.read_csv('ex1data2.txt', header=None, names=["Area","BedRooms","Price"])
```

## Calculating Mean/Standard Daviation

```python
rows=data.shape[0]

cols=len(data.columns)

X2=data.iloc[:,:cols-1]
Y2=data.iloc[:,cols-1]

area=data.iloc[:,0]
beds=data.iloc[:,1]
price=data.iloc[:,2]

areaMean=statistics.mean(area)
bedsMean=statistics.mean(beds)
priceMean=statistics.mean(price)

areaSTDev=statistics.stdev(area)
bedsSTDev=statistics.stdev(beds)
priceSTDev=statistics.stdev(price)
```

## Applying Normalization

```python
for i in range(rows):
    data.iloc[i,0]=(data.iloc[i,0]-areaMean)/areaSTDev
    data.iloc[i,1]=(data.iloc[i,1]-bedsMean)/bedsSTDev
    data.iloc[i,2]=(data.iloc[i,2]-priceMean)/priceSTDev
```

## Initializing Data

```python
thetas=[[ 0 for i in range(cols)]]

thetas=np.transpose(thetas)

print(thetas)

matrix=data.to_numpy()

X=np.ones((rows,cols))
Y=np.ones((rows,1))

X[:,1:]=matrix[:,:(cols-1)]
Y[:,0]=matrix[:,-1]
```

## Running Iterations

```python
alpha=0.001

print(thetas[0].shape)

for i in range(500):
    
    f1=np.transpose(np.subtract(np.dot(X,thetas),Y))
    f2=np.subtract(np.dot(X,thetas),Y)
    f3=np.dot(f1,f2)

    cost=(1/(2*rows))*f3
    
    print("Iteration #:",i)
    print(cost)
    
    for m in range(cols):
        thetas[m][0]=thetas[m][0]-((alpha/rows)*cost)
    
    print(thetas)
```

## Using Normal Equations

```python

x_transpose = np.transpose(X2)
thetas=np.linalg.pinv(x_transpose*X2)*x_transpose*Y2

print(thetas)
```
