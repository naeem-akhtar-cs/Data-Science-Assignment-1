## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

## Reading Data from File

```python
#Reading data from ex1data1.txt file
data=pd.read_csv('ex1data1.txt', header=None, names=["Population of City in 10,000s","Profit 10,000 $"])

rows=data.shape[0]
```

## Initializing Values

```python
#print("Number of Rows:", rows)

alpha=0.001

#Randomly Selected
theta_0_Prev=1
theta_1_Prev=1

val_0_Prev=5000
val_1_Prev=5000
```

## Running Iterations

```python
i=0
while True:
    
    val_0=0
    val_1=0
    
    for j in range(rows-1):
        
        #Predicted values
        predicted_0=theta_0_Prev+(theta_0_Prev*data.iloc[j,0])
        predicted_1=theta_1_Prev+(theta_1_Prev*data.iloc[j,0])
        
        
        val_0=val_0+(predicted_0-data.iloc[j,1])
        val_1=val_1+(((predicted_1-data.iloc[j,1]))*data.iloc[j,0])
        
        
    theta_0=theta_0_Prev - (alpha*(1/rows)*val_0)
    theta_1=theta_1_Prev - (alpha*(1/rows)*val_1)
    
    #cost.append(val_1)
    
    #print("Change: ",theta_0_Prev-theta_0)
    
    theta_0_Prev=theta_0
    theta_1_Prev=theta_1
    
    if (math.sqrt((val_0_Prev-val_0)*(val_0_Prev-val_0)))<=0.001 and (math.sqrt((val_1_Prev-val_1)*(val_1_Prev-val_1)))<=0.001:
        print("Iteration #:",i)       
        print(val_0_Prev,val_1_Prev)
        print(val_0,val_1)
        break
    
    i=i+1
    print(i,":",val_0,":",val_1)
    val_0_Prev=val_0
    val_1_Prev=val_1
```

## Initializing Values for Plot

```python
print(theta_0,theta_1)
    
#plt.plot([min(iterations), max(iterations)], [min(cost), max(cost)], color='blue')

#Plotting Data

X1 = min(data.iloc[:, 0])
X2 = max(data.iloc[:, 0])

Y_pred=theta_0+(theta_1*data.iloc[:, 0])

Y1=min(Y_pred)
Y2=max(Y_pred)
```

## Plotting Scatter Graph

```python
#Scatter Plot of Data
data.plot.scatter(x='Population of City in 10,000s', y='Profit 10,000 $', title= "Population and Profit Plot", color='red', marker='x')
            
plt.plot([X1, X2], [Y1, Y2], color='blue')

plt.legend(["Linear Regression", "Training Data"], loc=4)
```
